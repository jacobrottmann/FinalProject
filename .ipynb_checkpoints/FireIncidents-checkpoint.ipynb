```python
import pandas as pd
import numpy as np
fire_incidents = pd.read_csv("FireIncidents.csv")
fire_incidents.head()
```

```python
filtered = fire_incidents.filter(["type_description","neighborhood"])
filtered.head()
```

```python
grouped = filtered.groupby("neighborhood").count()
grouped.head()
```

```python
fewest_fires = grouped.nsmallest(10, "type_description")
fewest_fires.rename(columns = {'type_description' : 'Fires'}, inplace = True)
fewest_fires
```

```python
fewest_fires.plot(kind='bar')
```

Go back to the original dataframe, but only keep the important stuff (using drop instead of filter for some reason)

```python
simplified = fire_incidents.drop(["call_no","address","alarms","primary_unit","census_tract","council_district",
                                  "ward","tract","public_works_division","pli_division","police_zone",
                                  "fire_zone","latitude","longitude"], axis = 1)
simple = simplified.dropna()
simple.head()
```

Get rid of any fire not in 2018. Then get rid of alarm time, we don't need it anymore.

```python
simple = simple.loc[simple["alarm_time"].str.startswith("2018")]
recent_fires = simple.drop(["alarm_time"], axis=1)
recent_fires.head()
```

Get rid of everything that isn't a building fire, then get the number of fires in each neighborhood

```python
recent_fires = recent_fires[recent_fires["incident_type"] == 111]
neighborhood_sort = recent_fires.groupby("neighborhood").count()
least_fires = neighborhood_sort.nsmallest(25, "incident_type")
least_fires
```
